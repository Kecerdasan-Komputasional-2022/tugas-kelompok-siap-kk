```python
import tensorflow as tf
import pandas as pd
import csv
import numpy as np
```

```python
data_csv = pd.read_csv('./data/data.csv')
```

```python
bedrooms = data_csv["bedrooms"]
bathrooms = data_csv["bathrooms"]
luas = data_csv["sqft_living"]
floor = data_csv["floors"]
prices = data_csv["price"]
```

```python
rumah_all = []

for index in range(len(bedrooms)):
  rumah = []
  rumah.append(bedrooms[index])
  rumah.append(bathrooms[index])
  rumah.append(luas[index])
  rumah.append(floor[index])

  rumah_all.append(rumah)
```

```python
print(rumah_all)
rumah_all = np.array(rumah_all)
print(rumah_all)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units = 3, input_shape = [4]),
    # tf.keras.layers.Dense(units = 2, input_shape = [4]),
    # tf.keras.layers.Dense(units = 1)
])
```

```python
model.compile(
     loss = 'mae',
     optimizer = 'sgd',
     metrics = ['accuracy']
 )
```

```python
model.fit(
    rumah_all, prices, epochs = 500
)
```

```python
kamar_tidur = int(input("Masukan jumlah kamar tidur yang anda inginkan: "))
kamar_mandi = int(input("Banyak Kamar mandi: "))
luas_rumah = int(input("luas rumah anda?"))
jumlah_lantai = int(input("jumlah lantai rumah anda?"))
```

```python
prediksi_harga = model.predict([[kamar_tidur, kamar_mandi, luas_rumah, jumlah_lantai]])
print("Perkiraan harga: {} USD".format(prediksi_harga[0][0]))
```
